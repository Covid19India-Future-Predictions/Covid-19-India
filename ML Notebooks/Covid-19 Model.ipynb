```python
def date(a):
    st=''
    if a<92:
        st=str(a-60)+'-Mar-20'
        return st
    if a<122:
        st=str(a-91)+'-Apr-20'
        return st
    if a<153:
        st=str(a-121)+'-May-20'
        return st
    if a<184:
        st=str(a-152)+'-Jun-20'
        return st
    if a<214:
        st=str(a-183)+'-Jul-20'
        return st
```
