```python
from bs4 import BeautifulSoup
import re
```

```python
new=pd.read_csv(r'D:\D\prgrms\Confirmed3.csv')
```

```python
date=list(new['Date'])
date
```

```python
def clean1(da):
    da=BeautifulSoup(da,'lxml').get_text()
    da=re.sub(r'-Mar[A-Za-z0-9.-]+',' 3',da)
    da=re.sub(r'-A[A-Za-z0-9.-]+',' 4',da)
    da=re.sub(r'-May[A-Za-z0-9.-]+',' 5',da)
    da=re.sub(r'-J[A-Za-z0-9.-]+',' 6',da)
    mo=da.split()
    if mo[1]=='3':
        mo[0]=int(mo[0])+60
    if mo[1]=='4':
        mo[0]=int(mo[0])+91
    if mo[1]=='5':
        mo[0]=int(mo[0])+121
    if mo[1]=='6':
        mo[0]=int(mo[0])+152
    return mo[0]
```

```python
for i in range(len(date)):
    date[i]=clean1(date[i])
```

```python
new['Date']=date
```

```python
new.to_csv('D:/D/prgrms/Confirmed3.csv',index=False)
```
